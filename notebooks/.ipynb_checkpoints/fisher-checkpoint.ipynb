```julia
include("../src/celerite.jl")
include("../TransitModel/transit.jl")
using Plots
using Statistics
using ForwardDiff
using LaTeXStrings
using LinearAlgebra
Plots.scalefontsizes(0.75)
```

```julia
function trapezoidal_transit(t, t0, r, d, tin)
    trans = zeros(Real, length(t))
    t1 = (t0-d/2-tin)
    t2 = (t0-d/2)
    t3 = (t0+d/2)
    t4 = (t0+d/2+tin)
    intransit = (t.<=t3).&(t.>=t2)
    ingress = (t.<t2).&(t.>t1)
    egress = (t.<t4).&(t.>t3)
    trans[intransit] .= -(r^2)
    trans[ingress] .= -(t[ingress].-t1)*(r^2)/tin
    trans[egress] .= (1 - (r^2)).+(t[egress].-t3)*(r^2)/tin.-1
    return trans
end

function get_derivative(t, x0, p)
    if p == nothing
        d = transit(t, x0...)[1]
    else
        df = @eval (t, t0, b0, rp, d, u) -> ForwardDiff.derivative($p -> transit(t, t0, b0, rp, d, u)[1], $p)
        d = Base.invokelatest(df, t, x0...)
    end
    return d
end

function get_derivative_trapezoid(t, x0, p)
    if p == nothing
        d = transit(t, x0...)[1]
    else
        df = @eval (t, t0, rp, d, tin) -> ForwardDiff.derivative($p -> trapezoidal_transit(t, t0, rp, d, tin), $p)
        d = Base.invokelatest(df, t, x0...)
    end
    return d
end

function fisher(t, gp, df, log_wn=nothing)
    m = length(gp.Q[1,:])
    n = length(t)
    if log_wn != nothing
        if m > 1 
            wn_vec = zeros(n*m)
            df = kron(df, ones(m))
            for i in 1:m
                wn_vec[i:m:end] .= exp(log_wn[i]) 
            end
        else
            wn_vec = exp(log_wn)
        end
        celerite.compute!(gp, t, wn_vec)
    else
        if m > 1
            df = kron(df, ones(m))
        end
    end
    return df'*celerite.apply_inverse(gp, df)
end

function make_gp(t, log_s, log_w0, log_wn, ds0dlam=nothing, bins=nothing)
    if ds0dlam != nothing 
        b = diff(bins)
        wnfact = sqrt(sum(1 ./b))/length(b)
        log_wn = log_wn - log(wnfact)
        c = [(bins[i]+(ds0dlam-1)*bins[i-1])+0.5*ds0dlam*(bins[i]-bins[i-1])*(bins[i]-bins[i-1]+1) for i in 2:length(bins)]./bins[end]
        c[1] = 1
        for i in 2:length(c)
            c[i] = (c[i-1]+c[i])/c[1]
        end
        log_wn = [log_wn - 0.5*log(bins[i]-bins[i-1]) for i in 2:length(bins)]
        m = length(c)
        Q = broadcast(*, c, c')
        wn_vec = zeros(length(t)*m)
        for i in 1:m
            wn_vec[i:m:end] .= exp(log_wn[i]) 
        end
    else
        Q = ones(1, 1)
        c = nothing
    end
    log_q = log(1/sqrt(2))
    if ds0dlam != nothing
        gp = celerite.Celerite(celerite.SHOTerm(log_s, log_q, log_w0), Q)
        logD = celerite.compute!(gp, t, wn_vec)
    else
        gp = celerite.Celerite(celerite.SHOTerm(log_s, log_q, log_w0))
        logD = celerite.compute!(gp, t, exp.(log_wn))
    end
    return gp, c
end
```

```julia
t = collect(-4:0.01:4)

t0 = 0
b0 = 0
rp = 0.5
d = 1
tin = d/3
u = [0.5, 0.5]
#x0 = [t0, b0, rp, d, u]
x0 = [t0, rp, d, tin]

log_s = -16
w0T = 10
log_w0 = log(w0T) - log(d)
println(log_w0)
log_wn = -10
ds0dlam = [0.005, 0.01]#, 0.05, 0.1]

wn = collect(-12:0.1:0)
bins = [0, 100, 1000]

polyerr = zeros(length(wn), length(ds0dlam))
monoerr = zeros(length(wn), length(ds0dlam))
# should be exact variance for linear parameter rp 
df = get_derivative_trapezoid(t, x0, :rp)

for (i, w) in enumerate(wn)
    for (j, c) in enumerate(ds0dlam)
        polygp, scales = make_gp(t, log_s-log_w0, log_w0, w, c, bins)
        monogp, trash = make_gp(t, log_s-log_w0+log(mean(scales)), log_w0, w)
        polyerr[i,j] = 1/sqrt(fisher(t, polygp, df))
        monoerr[i,j] = 1/sqrt(fisher(t, monogp, df))
    end
end
```
